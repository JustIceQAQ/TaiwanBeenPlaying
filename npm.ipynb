```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import json
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
CheckPageInfo = True
CheckPageNum = 1
AllactivityList = []
while CheckPageInfo:
    req2 = requests.get(f'https://www.npm.gov.tw/zh-tw/JasonResult.aspx?lang=1&GetType=5&pageNo={CheckPageNum}&sno=03000060&Key=&at=&k=')
    CVC = json.loads(req2.text)['ResultList'][0]['data']
    if CVC:
        soup2 = bs(CVC, 'html5lib')
        for ii in soup2.findAll('li'):
            activityDict = {}
            activityDict['活動名稱'] = ii.find('h4').text
            activityDict['活動網址'] = "https://www.npm.gov.tw/zh-tw/" + ii.find('a')['href'].replace(' ', '') 
            multipleInfoList = ii.find('div', {"class":"words"}).findAll('b')
            for mil in multipleInfoList:
                if "展期：" in mil.text:
                    try:
                        activityDict['開始日期'], activityDict['結束日期'] = mil.text.replace("展期：", "").strip().split("~")
                    except ValueError:
                        activityDict['開始日期'] = mil.text.replace("展期：", "").strip().split("~")[0]
                        activityDict['結束日期'] = mil.text.replace("展期：", "").strip().split("~")[0]
                elif "陳列室：" in mil.text:
                    activityDict['展場'] = mil.text.replace("陳列室：", "").strip() 
                elif "時間：" in mil.text:
                    activityDict['時間'] = mil.text.replace("時間：", "").strip()
        #             activityTimeStart, activityTimeEnd, activityTimeRemarks = activityTime.split(",")[0]
                elif "展場：" in mil.text:
                    activityDict['展場'] = mil.text.replace("展場：", "").strip()
                elif "票價：" in mil.text:
                    activityDict['票價'] = mil.text.replace("票價：", "").strip()
            AllactivityList.append(activityDict)
        CheckPageNum+=1
    else:
        CheckPageInfo = False
```
